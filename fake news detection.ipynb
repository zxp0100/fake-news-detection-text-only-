```python
import re
import nltk
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from collections import Counter
from nltk.corpus import stopwords, wordnet
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.util import ngrams
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import PassiveAggressiveClassifier, LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Bidirectional, Dense, Dropout, Input, LSTM, Reshape
from wordcloud import WordCloud

# Download necessary NLTK data
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('vader_lexicon')


### FUNCTION USED ###
def print_label_statistics(df, dataset_name):
    """Prints the count and percentage of labels for a given dataset."""
    label_counts = df["label"].value_counts()
    label_percentage = (df["label"].value_counts(normalize=True) * 100).round(2)
    
    label_stats = pd.DataFrame({"Count": label_counts, "Percentage": label_percentage})
    
    print(f"{dataset_name} :")
    print(label_stats)
    print("\n") 

def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words("english"))
    tokens = [word for word in tokens if word not in stop_words]
    pos_tags = pos_tag(tokens)
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word, get_wordnet_pos(tag)) for word, tag in pos_tags]
    tokens = [word for word in tokens if len(word) > 2]
    return " ".join(tokens)

def get_top_ngrams(texts, n=1, top_k=10):
    """Function to extract top n-grams (unigram or bigram)."""
    ngram_list = []
    for text in texts:
        tokens = text.split()
        ngram_list.extend(ngrams(tokens, n))
    
    ngram_freq = Counter(ngram_list).most_common(top_k)
    return ngram_freq

def build_bilstm_model(input_dim):
    model = Sequential([
        Input(shape=(input_dim,)),
        Dense(256, activation='relu'),
        Dropout(0.3),
        Dense(128, activation='relu'),
        Reshape((128, 1)),
        Bidirectional(LSTM(64, return_sequences=True)),
        Dropout(0.3),
        Bidirectional(LSTM(32)),
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(1, activation='sigmoid')
    ])
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model    

def plot_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['False', 'True'], yticklabels=['False', 'True'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(title)
    plt.show()


# Read datasets
df = pd.read_excel(r"C:\Users\pzx12\OneDrive\Desktop\FYP\dataset\liar_combineddataset.xlsx", header=None)

# Select only the relevant columns
df = df[[2, 1]]
df.columns = ["statement", "label"]

# Convert multi-class labels to binary
df["label"] = df["label"].replace({"half-true": True, "mostly-true": True, "pants-fire": False, "barely-true": False})

# Check the distribution of the dataset
print_label_statistics(df, "LIAR")

# Apply text preprocessing function to the 'statement' column
df["clean_text"] = df["statement"].apply(preprocess_text)

# Sentiment Analysis using VADER
sia = SentimentIntensityAnalyzer()
df["sentiment"] = df["statement"].apply(lambda x: sia.polarity_scores(str(x))["compound"] + 1)


## BASIC STATISTICS FOR 'clean_text' COLUMN ##
# Word Count
df["word_count"] = df["clean_text"].apply(lambda x: len(x.split()))
df["unique_word_count"] = df["clean_text"].apply(lambda x: len(set(x.split())))

print("WordCount Statistics :")
print(df[["word_count", "unique_word_count"]].describe().round(0).astype(int), "\n")


# Get top unigrams and bigrams
top_unigrams = get_top_ngrams(df["clean_text"], n=1, top_k=10)
top_bigrams = get_top_ngrams(df["clean_text"], n=2, top_k=10)

# Visualization of unigrams and bigrams
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Unigram Plot
unigram_words, unigram_freqs = zip(*top_unigrams)
sns.barplot(x=list(unigram_freqs), y=[" ".join(word) for word in unigram_words], ax=axes[0], palette="viridis")
axes[0].set_title("Top 10 Unigrams")
axes[0].set_xlabel("Frequency")

# Display values on bars
for container in axes[0].containers:
    axes[0].bar_label(container, fmt="%d", padding=3)

# Bigram Plot
bigram_words, bigram_freqs = zip(*top_bigrams)
sns.barplot(x=list(bigram_freqs), y=[" ".join(word) for word in bigram_words], ax=axes[1], palette="magma")
axes[1].set_title("Top 10 Bigrams")
axes[1].set_xlabel("Frequency")

# Display values on bars
for container in axes[1].containers:
    axes[1].bar_label(container, fmt="%d", padding=3)

plt.tight_layout()
plt.show()


## BASIC SENTIMENT ANALYSIS ##
# Overall statistics 
print("Overall Sentiment Score:")
print(df["sentiment"].describe())
print("Mode:", df["sentiment"].mode().values)

# Statistic based on each label
df_true = df[df["label"] == True]
df_false = df[df["label"] == False]

print("\nSentiment Score for 'True' :")
print(df_true["sentiment"].describe())
print("Mode:", df_true["sentiment"].mode().values)

print("\nSentiment Score for 'False' :")
print(df_false["sentiment"].describe())
print("Mode:", df_false["sentiment"].mode().values)


# TF-IDF Vectorization
vectorizer = TfidfVectorizer(max_features=10000, ngram_range=(1, 2))
X_tfidf = vectorizer.fit_transform(df["clean_text"])
X_tfidf_df = pd.DataFrame(X_tfidf.toarray())
X_tfidf_df.columns = X_tfidf_df.columns.astype(str)


## TOP 1O UNIGRAM & BIGRAM (HIGHEST IN AVERAGE TF-IDF SCORES) FOR TRUE & FALSE STATEMENTS ##
# Get the feature names (tokens) from the vectorizer
feature_names = vectorizer.get_feature_names_out()

# Create a DataFrame with the TF-IDF values and feature names
X_tfidf_df2 = pd.DataFrame(X_tfidf.toarray(), columns=feature_names)
X_tfidf_df2["label"] = df["label"].values

# Filter out rows where the label is True/False and drop the label column
df_true2 = X_tfidf_df2[X_tfidf_df2["label"] == True].drop(columns="label")
false_df2 = X_tfidf_df2[X_tfidf_df2["label"] == False].drop(columns="label")

# Compute mean TF-IDF score for each token
true_means = df_true2.mean().sort_values(ascending=False)
false_means = false_df2.mean().sort_values(ascending=False)

# Separate unigrams and bigrams
unigrams_true = true_means[true_means.index.map(lambda x: len(x.split()) == 1)].head(10)
bigrams_true = true_means[true_means.index.map(lambda x: len(x.split()) == 2)].head(10)

unigrams_false = false_means[false_means.index.map(lambda x: len(x.split()) == 1)].head(10)
bigrams_false = false_means[false_means.index.map(lambda x: len(x.split()) == 2)].head(10)

# Display
print("\nTop 10 Unigrams for TRUE Statements:")
print(unigrams_true)

print("\nTop 10 Bigrams for TRUE Statements:")
print(bigrams_true)

print("\nTop 10 Unigrams for FALSE Statements:")
print(unigrams_false)

print("\nTop 10 Bigrams for FALSE Statements:")
print(bigrams_false)


# Extract sentiment data
X_sentiment = df[["sentiment"]].reset_index(drop=True)

# Create a dictionary containing different feature sets
X_sets = {
    "TF-IDF": X_tfidf_df,
    "TF-IDF + Sentiment": pd.concat([X_tfidf_df, X_sentiment], axis=1)
}


## Train and evaluate individual machine learning models and ensemble models (hard and soft voting), with or without sentiment scores  
for set_name, X_final in X_sets.items():
    print(f"Evaluating feature set: {set_name}\n")
    X_train, X_test, y_train, y_test = train_test_split(X_final, df['label'], test_size=0.2, random_state=42, stratify=df['label'])

    # Initialize traditional machine learning models
    rf = RandomForestClassifier(n_estimators=100, random_state=42)
    pa = PassiveAggressiveClassifier(max_iter=1000, random_state=42)
    svm = SVC(kernel='linear', probability=True, random_state=42)
    lr = LogisticRegression(max_iter=1000, random_state=42)
    nb = MultinomialNB()
    
    models = [("Random Forest", rf), ("PA Classifier", pa), ("SVM", svm), ("Logistic Regression", lr), ("Naive Bayes", nb)]
    
    predictions, probabilities = [], []
    for name, model in models:
        model.fit(X_train, y_train)    
        y_pred = model.predict(X_test)
        plot_confusion_matrix(y_test, y_pred, title=f"Confusion Matrix - {name}")
        predictions.append(y_pred)
        if hasattr(model, "predict_proba"):
            probabilities.append(model.predict_proba(X_test)[:, 1])
        elif hasattr(model, "decision_function"):
            probabilities.append((model.decision_function(X_test) - model.decision_function(X_test).min()) / 
                (model.decision_function(X_test).max() - model.decision_function(X_test).min()))
        print(f"\nClassification Report - {name}:")
        print(classification_report(y_test, y_pred))

    
    # Train a Bi-LSTM model
    bilstm = build_bilstm_model(X_train.shape[1])
    bilstm.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))
    bilstm_preds = (bilstm.predict(X_test) > 0.5).astype("int32").flatten()
    plot_confusion_matrix(y_test, bilstm_preds, title="Confusion Matrix - BI-LSTM")
    bilstm_probs = bilstm.predict(X_test).flatten()
    
    print("\nClassification Report - BI-LSTM:")
    print(classification_report(y_test, bilstm_preds))

    
    # Hard voting ensemble
    hard_voting_preds = np.array(predictions + [bilstm_preds]).T
    final_hard_preds = [np.bincount(row).argmax() for row in hard_voting_preds]

    # Soft voting ensemble
    final_soft_preds = (np.column_stack(probabilities + [bilstm_probs]).mean(axis=1) > 0.5).astype(int)

    plot_confusion_matrix(y_test, final_hard_preds, title="Confusion Matrix - Hard Voting")
    print("\nClassification Report - Hard Voting Ensemble:")
    print(classification_report(y_test, final_hard_preds))
    plot_confusion_matrix(y_test, final_soft_preds, title="Confusion Matrix - Soft Voting")
    print("\nClassification Report - Soft Voting Ensemble:")
    print(classification_report(y_test, final_soft_preds))
```
